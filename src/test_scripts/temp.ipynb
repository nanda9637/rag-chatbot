```python
import os
from tavily import TavilyClient
from dotenv import load_dotenv

load_dotenv()

# Ensure your API key is set as an environment variable named TAVILY_API_KEY
# If not set, you can pass it directly: tavily_client = TavilyClient(api_key="tvly-YOUR_API_KEY")
tavily_client = TavilyClient(api_key=os.environ.get("TAVILY_API_KEY"))

query = "What is the today's date?"

# Execute a basic search query
qna_response = tavily_client.qna_search(
    query=query,
    include_answer=True,
    include_raw_content=False,
    max_results=2,      
)

# Print the full response (which is a dictionary)
print(qna_response)

# To get a direct answer, use the qna_search method
# answer = tavily_client.qna_search(query)
# print(answer)
```

```python
import os
from dotenv import load_dotenv, find_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 1. Load the API key from the .env file
load_dotenv(find_dotenv())
groq_api_key = os.getenv("GROQ_API_KEY")

if not groq_api_key:
    raise ValueError("GROQ_API_KEY environment variable not set")


model_names = ["mixtral-8x7b-32768", "llama-3.3-70b-versatile", "gpt-4o-mini"]
# 2. Initialize the Groq LLM
# You can specify the model, e.g., "llama3-8b-8192" or "llama3-70b-8192"
llm = ChatGroq(
    api_key=groq_api_key,
    model_name="llama-3.3-70b-versatile" 
)

# 3. Define a prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that responds concisely."),
    ("user", "{input}")
])

# 4. Define an output parser
output_parser = StrOutputParser()

# 5. Create a chain: prompt | model | output_parser
chain = prompt | llm | output_parser

# 6. Invoke the chain with a user input
user_input = "What is the impact of Agentic AI on traditional jobs? Provide a brief summary in exactly 10 words."
response = chain.invoke({"input": user_input})

print(response)
```
